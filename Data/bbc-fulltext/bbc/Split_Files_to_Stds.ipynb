```python
def createShuffelist(filepath, number_files, number_group):
    from random import shuffle
    from glob import glob
    i = 1
    ls = []

    l_df =[]
    
    for root, directory, files in os.walk(filepath):
        for file in files:
            filepath = os.path.join(root,file)
            l_df.append(file)
    df = pd.DataFrame(l_df)
    
    while i <= number_group:
        ls.append((df.sample(number_files)).values)
        i = i + 1
    return ls
```

```python
def rename_file_in_folder(folderpath):
    import os, sys

    files = os.listdir(folderpath)
    counter = 0

    for file in files:
        if file[-4:] == '.txt':
            print("oldname: " + str(file) + "::: newname: " + str("{0}.txt".format(counter)))
            os.rename(folderpath + file, folderpath + "{0}.txt".format(counter))
            counter += 1
```

```python
import os
import shutil


## Get std list
std_list = pd.read_csv('ls.csv',sep=',', header=None)

## Get shuffer list of files
filepath = './tech/'

ls = createShuffelist(filepath, number_files=50, number_group=len(std_list))
print('size of class:' + str(len(ls)) + " Stds")

## Copy files
i = 0
for file_name_list in ls:
    for file_name in file_name_list:
        full_file_name = os.path.join(filepath, file_name[0])
        print('coppy file: ' + str(full_file_name))

        destname = str(std_list.iloc[i,0])+"_"+str(std_list.iloc[i,1])
        
        dest = filepath + "/" + str(destname) + "/"
        if not os.path.exists(dest):
            os.makedirs(dest)

        if (os.path.isfile(full_file_name)):
            shutil.copy(full_file_name,dest )
    i = i +1
```

```python
## Change name of files in each folder
print("rename!")
i = 0
while i <= len(std_list):
    destname = str(std_list.iloc[i,0])+"_"+str(std_list.iloc[i,1])
    dest = filepath + "/" + str(destname) + "/"
    rename_file_in_folder(dest)
    i = i +1
```

```python
import os
import zipfile

path = "D:\git\intro_text_analytics\DATA\DistributedFolder"
path = os.path.abspath(os.path.normpath(os.path.expanduser(path)))
for folder in os.listdir(path):
    zipf = zipfile.ZipFile('{0}.zip'.format(os.path.join(path, folder)), 'w', zipfile.ZIP_DEFLATED)
    for root, dirs, files in os.walk(os.path.join(path, folder)):
        for filename in files:
            zipf.write(os.path.abspath(os.path.join(root, filename)), arcname=filename)
    zipf.close()
```

```python
import os
path = "./"
path = os.path.abspath(os.path.normpath(os.path.expanduser(path)))
print(path)
```

```python
import pandas as pd
data_path = "./"
df_raw = pd.DataFrame([],columns=['filename','content','category'])

for root, directories, files in os.walk(data_path):
    for filename in files:
        classname = (os.path.basename(root))
        if (classname in ['business', 'entertainment', 'politics', 'sport', 'tech']):
            _name = os.path.splitext(filename)[0]
            filepath = os.path.join(root, filename)
            f = open(filepath,"r",encoding='latin-1') 
            content = f.read()
            content = " ".join(str(content).split())
#             content = str(content).replace("\n"," ")
            pieces = {'filename': _name, 'content': (content), 'category':classname}
            df_raw = df_raw.append(pieces,ignore_index=True)
df_raw
```

```python
df_raw.shape
```

```python
df_raw = df_raw.drop('filename',1)
df_raw
```

```python
from sklearn.model_selection import train_test_split
trainingSet, testSet = train_test_split(df_raw, test_size=0.3)
```

```python
trainingSet.to_csv('trainingset.csv',sep='^',index=None,header=True)
testSet.to_csv('test_groundtruth.csv',sep='^',index=None,header=True)
testdata = testSet['content']
testdata.to_csv('testdata.csv',sep='^',index=None,header=True)
```
